```python
import requests
from bs4 import BeautifulSoup
import os

base_url_oromo = "https://www.bible.com/bible/3202/"
base_url_dholuo = "https://www.bible.com/bible/2652/"

books = [
    "GEN", "EXO", "LEV", "NUM", "DEU", "JOS", "JDG", "RUT", "1SA", "2SA",
    "1KI", "2KI", "1CH", "2CH", "EZR", "NEH", "EST", "JOB", "PSA", "PRO",
    "ECC", "SNG", "ISA", "JER", "LAM", "EZK", "DAN", "HOS", "JOL", "AMO",
    "OBA", "JON", "MIC", "NAM", "HAB", "ZEP", "HAG", "ZEC", "MAL", "MAT",
    "MRK", "LUK", "JHN", "ACT", "ROM", "1CO", "2CO", "GAL", "EPH", "PHP",
    "COL", "1TH", "2TH", "1TI", "2TI", "TIT", "PHM", "HEB", "JAS", "1PE",
    "2PE", "1JN", "2JN", "3JN", "JUD", "REV"
]


output_dir_oromo = "lit/oromo"
output_dir_dholuo = "lit/dholuo"
os.makedirs(output_dir_oromo, exist_ok=True)
os.makedirs(output_dir_dholuo, exist_ok=True)

def save_chapter(book, chapter_num, content, output_dir):
    filename = f"{book}.txt"
    with open(os.path.join(output_dir, filename), "a", encoding="utf-8") as file:
        file.write(content + "\n")

for lang in ['dholuo', 'oromo']:
  for book in books:
      for chapter_num in range(1, 51):
          if lang == 'dholuo':
            chapter_url = f"{base_url_dholuo}/{book}.{chapter_num}.MLM"
          else:
            chapter_url = f"{base_url_oromo}/{book}.{chapter_num}.MACQUL"
          response = requests.get(chapter_url)
          soup = BeautifulSoup(response.content, "html.parser")
          chapter_content = soup.find_all("span", class_="ChapterContent_content__RrUqA")
          chapter_text = ""
          for sp in chapter_content:
            chapter_text += sp.get_text(separator="\n")
          if chapter_text:
            if lang == 'dholuo':
              save_chapter(book, chapter_num, chapter_text, output_dir_dholuo)
            else:
              save_chapter(book, chapter_num, chapter_text, output_dir_oromo)
```

```python
import os

def pre_process(dir):
  dholuo_stop_words = list("an in wan en un gin mara magi marwa magu mari margi mar ma magi macha maka ka kacha gir gira nade adi wan wa ooyo kamano ok".split())
  oromo_stop_words = list("na ani an naa naaf natti narraa koo kiyya too tiyya si ati sii siif sitti siin sirraa kee isa isaa isaatii isarraa isii ishii isee ishee isiin ishii ishiif ishiitti ishiin ishiirraa nu nuti nuy nu nuun nutti nuuf nurraa keenna keenya isini isinii isiniif isinitti isiniin isinirraa keesan isaan isaani isaani isaaniif isaanitti tiin isaanirraa akka itti garuu f kan ala tti sana tana kana sun fi inni ana akka ishee isaan nuti wal ni hin ani ise".split())
  for folder in os.listdir(dir):
      for file in os.listdir(f"{dir}/{folder}"):
          if file != ".ipynb_checkpoints":
              with open(f"{dir}/{folder}/{file}", "r") as f:
                  text = f.read()
              text = text.lower()
              text = ''.join(i for i in text if i == " " or "a" <= i <= "z")
              word_tokens = list(text.split())
              if folder == "oromo":
                  filtered_text = [word for word in word_tokens if (len(word) > 2 and word not in oromo_stop_words)]
              else:
                  filtered_text = [word for word in word_tokens if (len(word) > 2 and word not in dholuo_stop_words)]
              with open(f"{dir}/{folder}/{file}", "w") as f:
                  f.write(' '.join(filtered_text))
```

```python
import pandas as pd

def ds(a):
  df = pd.DataFrame(a)
  df.drop_duplicates(inplace=True)
  return df.values.tolist()
```

```python
def get_3grams(text):
    words = text.split()
    return [words[i:i+3] for i in range(len(words) - 2)]

ngrams_dholuo = get_3grams(open("/content/lit/all_dholuo.txt").readline())
ngrams_oromo = get_3grams(open("/content/lit/all_oromo.txt").readline())

ngrams_dholuo_gpt = get_3grams(open("/content/gpt/all_dholuo_gpt.txt").readline())
ngrams_oromo_gpt = get_3grams(open("/content/gpt/all_oromo_gpt.txt").readline())
```

```python
#Word2Vec

def get_w2v_dataset(model, ngrams):
    vectors = []
    for ngram in ngrams:
        vec = np.concatenate([model.wv[word] for word in ngram])
        vectors.append(vec)
    return vectors

w2v_dataset_dholuo = ds(get_w2v_dataset(model_dholuo, ngrams_dholuo))
w2v_dataset_oromo = ds(get_w2v_dataset(model_oromo, ngrams_oromo))

w2v_dataset_dholuo_gpt = ds(get_w2v_dataset(model_dholuo_gpt, ngrams_dholuo_gpt))
w2v_dataset_oromo_gpt = ds(get_w2v_dataset(model_oromo_gpt, ngrams_oromo_gpt))
```

```python
#SVD

def get_svd_dataset(vectorizer, u, s, ngrams):
    vectors = []
    for ngram in ngrams:
        idx = []
        for word in ngram:
            idx.append(vectorizer.vocabulary_[word])
        vec = np.concatenate([u[i] * s for i in idx])
        vectors.append(vec)
    return vectors

svd_dataset_dholuo = ds(get_svd_dataset(vectorizer_dholuo, svd_u_dholuo, svd_s_dholuo, ngrams_dholuo))
svd_dataset_oromo = ds(get_svd_dataset(vectorizer_oromo, svd_u_oromo, svd_s_oromo, ngrams_oromo))

svd_dataset_dholuo_gpt = ds(get_svd_dataset(vectorizer_dholuo_gpt, svd_u_dholuo_gpt, svd_s_dholuo_gpt, ngrams_dholuo_gpt))
svd_dataset_oromo_gpt = ds(get_svd_dataset(vectorizer_oromo_gpt, svd_u_oromo_gpt, svd_s_oromo_gpt, ngrams_oromo_gpt))
```

```python
#selection of hyperparameter

from Wishart import Wishart
from sklearn.metrics import calinski_harabasz_score as chs
import matplotlib.pyplot as plt

sig = 0.05
parameter = [i for i in range(10, 251, 5)]
objects = [np.array(obj)[:5000] for obj in [svd_dataset_dholuo, svd_dataset_dholuo_gpt, \
           w2v_dataset_dholuo, w2v_dataset_dholuo_gpt, \
           svd_dataset_oromo, svd_dataset_oromo_gpt, \
           w2v_dataset_oromo, w2v_dataset_oromo_gpt]]
ch = [[] for i in range(len(objects))]


for i in range(len(objects)):
  for k in parameter:
    wishart = Wishart(significance_level = sig, wishart_neighbors = k)
    predict = wishart.fit(objects[i])

    max_cluster = np.max(predict)
    min_cluster = np.min(predict)

    clusters = [[] for i in range(max_cluster - min_cluster + 1)]
    ch[i].append(chs(objects[i], predict) if len(clusters) > 1 else 0)
```

```python
def build_clusters(objects, clustering):
  max_cluster = np.max(clustering)
  min_cluster = np.min(clustering)
  clusters = [[] for i in range(max_cluster - min_cluster + 1)]
  for obj, cluster in zip(objects, clustering):
        clusters[cluster - min_cluster].append(obj)
  return clusters
```

```python
import random
from scipy.spatial.distance import pdist

def get_metrics(clusters):
  metrics = {"len":[], "len_maxnorm":[], "len_allnorm":[], \
             "cdist_max":[], "cdist_mean":[], \
             "pdist_max":[], "pdist_mean":[]}

  for cluster in clusters:
    n = len(cluster)
    metrics["len"].append(n)

    centroid = np.mean(cluster, axis=0)
    s = cdist_max = 0
    for vector in cluster:
      cdist_max = max(cdist_max, np.linalg.norm(vector - centroid))
      s += np.linalg.norm(vector - centroid)
    metrics["cdist_max"].append(cdist_max)
    metrics["cdist_mean"].append(s/n)

  for k in range(len(clusters)):
    metrics["len_maxnorm"].append(len(clusters[k])/max(metrics["len"]))
    metrics["len_allnorm"].append(len(clusters[k])/sum(metrics["len"]))
    d = d_max = 0
    if len(clusters) < 12:
      y = pdist(clusters[k])
      metrics["pdist_max"].append(max(y))
      metrics["pdist_mean"].append(2 * sum(y)/(len(clusters[k]) * (len(clusters[k]) - 1)))
    else:
      nn = int(metrics["len_allnorm"][k] * 5000)
      cl = random.sample(clusters[k], nn)
      for i in range(nn - 1):
        for j in range(i + 1, nn):
          dd = np.linalg.norm(np.array(cl[i]) - np.array(cl[j]))
          d += dd
          d_max = max(d_max, dd)
      metrics["pdist_max"].append(d_max)
      metrics["pdist_mean"].append(2 * d/(n * (n - 1)))

  return metrics
```

```python
from scipy.stats import ranksums

def wilcxon(a,b):
  for m in ["len", "len_maxnorm", "len_allnorm", \
             "cdist_max", "cdist_mean", \
             "pdist_max", "pdist_mean"]:
    p = ranksums(a[m],b[m]).pvalue
    print(m, p, p >= 0.05)
  print('---------------')
```
